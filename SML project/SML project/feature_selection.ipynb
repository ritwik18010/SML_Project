```
import pandas as pd
df=pd.read_csv('gdrive/My Drive/final_data.csv',dtype=dtypes,nrows=1000000)
```

```
df1=df.drop(columns=['Unnamed: 0'])
```

```
cat_feat=[]
for col in df1.columns:
  if dtypes[col]=='category':
    cat_feat.append(col)
```

```
df2=df1.drop(columns=cat_feat)
df2=df2.drop(columns=['HasDetections'])
```

```
import pandas as pd
import numpy as np


X = df2.iloc[:,:]  #independent columns
y = df1.iloc[:,-1]    #target column i.e price range
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```
