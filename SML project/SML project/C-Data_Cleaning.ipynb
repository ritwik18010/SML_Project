# Checking Type, Missing Value,Unique Values and SkewFactor for each Feature in Training Data

```
Column_Info = []
for col in Train.columns:
    Column_Info.append((col, Train[col].dtype, Train[col].nunique(), Train[col].isnull().sum() * 100 / Train.shape[0], Train[col].value_counts(normalize=True, dropna=False).values[0] * 100))
    
Columns_Status = pd.DataFrame(Column_Info, columns=['Feature', "DataType", 'Number of Unique Values', 'Percentage of Missing Values', 'Skewness Factor'])
Columns_Status.sort_values('Percentage of Missing Values', ascending=False)
```

# Removing of Features with High NA Rate and SkewFactor

```
Useful_Col = list(Train.columns)

for col in Train.columns:
    
    NA_Rate = Train[col].isnull().sum() / Train.shape[0]
    Skewed_Factor = Train[col].value_counts(normalize=True, dropna=False).values[0]
    
    NA_Rate_Threshold=0.7
    Skewed_Factor_Threshold=0.9
    
    
    # remove columns with high NA rate
    if NA_Rate > NA_Rate_Threshold:
        Useful_Col.remove(col)
    
    # remove columns with high Unbalanced values rate
    elif Skewed_Factor > Skewed_Factor_Threshold :
       Useful_Col.remove(col)
```

# Handling NaN Values in Numerical Features

```
import numpy as np
import statistics
def Outlier(Data):
  outlier=[]
  mean = statistics.mean(Data)
  
  standard_deviation = statistics.stdev(Data)
  
  for i in range(len(Data)):
    if Data[i]<(mean - 3* standard_deviation) or Data[i]>(mean + 3 * standard_deviation):
      outlier.append(Data[i])
  
  return outlier
# arr=[1,3,4,5,15,10,12,12,13,12,11,14,13,15,10,10,10,100,12,14,13,100, 12,10, 10,11,12,15,12,13,12,11,14,13,15,10,15,12,10,14,13,15,10]
# arr=np.array(arr)
# y=Outlier(arr)
# print(y)
```

```
import copy 
Training_Data=copy.deepcopy(Training_Set)
```

```
for col in Numerical_Features:
  data=Training_Set[col]
  data=list(data)
  list1=[]

  for i in range(len(data)):
    if not np.isnan(data[i]):
      list1.append(data[i])

  s_data=0
  s_data=sum(list1)

  Outlier_list=Outlier(list1)

  s_outlier=0
  s_outlier=sum(Outlier_list)

  diff_sum=s_data-s_outlier
  # c=0
  # for i in range(len(list1)):
  #   if(list1!='NaN'):
  #     c=c+1
  value=diff_sum/(len(list1)-len(Outlier_list))
  
  
  for i in range(len(data)):
    if np.isnan(data[i]):
      data[i]=value
  
  data=pd.DataFrame(data)
  Training_Data[col]=data
```
